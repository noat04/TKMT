Bạn hãy phát sinh ngẫu nhiên mảng ndarray có tên `arr` gồm 100 phần tử số nguyên có giá trị từ 0..1000. Thực hiện các thao tác sau:

### 1. Hiển thị các phần tử thỏa yêu cầu dưới đây

- 10 phần tử đầu tiên
- 5 phần tử cuối cùng
- Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100

```python
# Phát sinh ngẫu nhiên mảng arr gồm 100 phần tử số nguyên từ 0 đến 1000
import numpy as np
arr = np.random.randint(0, 1001, 100)

# 10 phần tử đầu tiên
first_10 = arr[:10]

# 5 phần tử cuối cùng
last_5 = arr[-5:]

# Các phần tử chia hết cho 5 và lớn hơn 100
chiahet5vs10 = arr[(arr % 5 == 0) & (arr > 100)]

first_10, last_5, chiahet5vs10
```

### 2. Cho biết mảng trên có bao nhiêu phần tử có giá trị nhỏ hơn 100. Hiển thị các phần tử đó

```python
less_than_100 = arr[arr < 100]

# Số lượng các phần tử nhỏ hơn 100
count_less_than_100 = len(less_than_100)

count_less_than_100
```

### 3. Thực hiện thống kê:

- Tính tổng mảng
- Tìm giá trị trung bình
- Tìm giá trị trung vị
- Tìm Q1, Q2, Q3
- Tìm phân vị 20 (percentile 20)

```python
# Tính tổng mảng
total_sum = np.sum(arr)

# Tính giá trị trung bình
mean_value = np.mean(arr)

# Tính giá trị trung vị
median_value = np.median(arr)

# Tính các phân vị Q1, Q2, Q3
q1 = np.percentile(arr, 25)  # Q1 (25th percentile)
q2 = np.percentile(arr, 50)  # Q2 (median, 50th percentile)
q3 = np.percentile(arr, 75)  # Q3 (75th percentile)

# Tính phân vị 20 (20th percentile)
percentile_20 = np.percentile(arr, 20)

total_sum, mean_value, median_value, q1, q2, q3, percentile_20
```

### 4. Thực hiện việc chọn ngẫu nhiên (không thay thế) 10 phần tử từ mảng trên 5 lần. Hiển thị kết quả lựa chọn mỗi lần

```python
# Chọn ngẫu nhiên 10 phần tử từ mảng, không thay thế, 5 lần
random_selections = [np.random.choice(arr, 10, replace=False) for _ in range(5)]

random_selections
```

### 5. Tìm phần tử xuất hiện nhiều nhất trong mảng

```python
# Tìm phần tử xuất hiện nhiều nhất trong mảng
unique_elements, counts = np.unique(arr, return_counts=True)
most_frequent_element = unique_elements[np.argmax(counts)]
frequency = counts[np.argmax(counts)]

most_frequent_element, frequency
```

### Câu 10: Kết hợp trên cùng một đồ thị

Sử dụng hàm `np.random.normal(size=10000)` để phát sinh 10000 số ngẫu nghiên thuộc phân phối chuẩn. Vẽ hình thể hiện phân phối như sau:

Hướng dẫn: thêm tham số `vert=False` trong hàm vẽ `boxplot()`

```python
# This cell contains output. Don't remove!
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Phát sinh 10000 số ngẫu nhiên từ phân phối chuẩn
data = np.random.normal(size=10000)

# Tạo một figure và 2 subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Vẽ histogram
axs[0].hist(data, bins=30, density=True, color='purple', alpha=0.6)
axs[0].set_title('Histogram')
axs[0].set_xlabel('Giá trị')
axs[0].set_ylabel('Mật độ')

# Vẽ boxplot theo chiều ngang
axs[1].boxplot(data, vert=False)
axs[1].set_title('Boxplot')
axs[1].set_xlabel('Giá trị')

# Hiển thị đồ thị
plt.suptitle('Đồ thị phân phối với Histogram và Boxplot')
plt.tight_layout()
plt.show()
```
