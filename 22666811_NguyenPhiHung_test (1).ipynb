### Câu 1 (1.5đ): Tính xác suất có ít nhất 30 người thực hiện giao dịch mua hàng

```python
#phan2
#1.	(1.5đ): Một trang web thương mại điện tử có tỷ lệ chuyển đổi (conversion rate) là 8%.
#Nếu có 300 khách truy cập trang web trong một ngày, tính xác suất có ít nhất 30 người thực hiện giao dịch mua hàng.

from scipy.stats import binom

# Tham số phân phối nhị thức
n = 300  # số sản phẩm
p = 0.08  # xác suất bị lỗi

# Tính xác suất P(X < 30)
x = binom.cdf(29, n, p)

# Tính xác suất P(X >= 30)
y = 1 - x
# In kết quả làm tròn 4 chữ số thập phân
print(f"Xác suất để it nhat 30 nguoi mua hang {y:.4f}")
```
