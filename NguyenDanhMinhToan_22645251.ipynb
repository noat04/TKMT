```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('thoi_tiet_khong_khi_vn_2023.csv')
```

```python
# Cho biết thông tin về các cột dữ liệu và kiểu của từng cột trong bảng
data.info()
```

```python
# Cho biết số lượng bản ghi bị thiếu giá trị cho các cột: Nhiệt_độ, Lượng_mưa và AQI.
print('Số Cho biết số lượng bản ghi bị thiếu giá trị cho cột Nhiệt_độ: ', data['Nhiệt_độ'].isnull().sum())
print('Số Cho biết số lượng bản ghi bị thiếu giá trị cho cột Lượng_mưa: ', data['Lượng_mưa'].isnull().sum())
print('Số Cho biết số lượng bản ghi bị thiếu giá trị cho cột AQI: ', data['AQI'].isnull().sum())
```

```python
# 1.	(0.75đ) Trung bình (mean), trung vị (median) và mode của nhiệt độ.
# 2.	(0.75đ) Độ lệch chuẩn (standard deviation) và phương sai (variance).
# 3.	(0.75đ) Nhiệt độ thấp nhất, cao nhất và biên độ nhiệt (range).
print('Trung bình: ', data['Nhiệt_độ'].mean())
print('Trung vị: ', data['Nhiệt_độ'].median())
print('Mode: ', data['Nhiệt_độ'].mode())
print('Độ lệch chuẩn: ', data['Nhiệt_độ'].std())
print('Phương sai: ', data['Nhiệt_độ'].var())
print('Nhiệt độ thấp nhất: ', data['Nhiệt_độ'].min())
print('Nhiệt độ lớn nhất: ', data['Nhiệt_độ'].max())
print('Range: ', data['Nhiệt_độ'].max() - data['Nhiệt_độ'].min())
```

```python
# Đếm số ngày có thời tiết "Nắng" và số ngày có thời tiết "Mưa to" hoặc "Dông".
data['Thời_tiết'].value_counts()
```

Vậy số ngày thời tiết "Nắng" :  1461 ngày
Vậy số ngày thời tiết "Mưa to" : 403 ngày
Vậy số ngày thời tiết "Dông" : 412 ngày


```python
# Vẽ đồ thị Pie để so sánh phân bố các loại thời tiết trong năm
data['Thời_tiết'].value_counts().plot(kind='pie', autopct='%.2f%%')
```

```python
# Liệt kê 20 ngày có chỉ số AQI cao nhất và thành phố tương ứng
data.sort_values(by='AQI', ascending=False).head(20)
```

```python
#Cho biết thành phố có lượng mưa trung bình năm cao nhất.
thanhphotop = data.sort_values(by='Lượng_mưa', ascending=False).max()
print('thành phố có lượng mưa trung bình năm cao nhất:',thanhphotop['Thành_phố'])
thanhphotop = data.sort_values(by='Lượng_mưa', ascending=False).iloc[0]
print('Thành phố có lượng mưa trung bình năm cao nhất:', thanhphotop['Thành_phố'])
```

```python
# Tính nhiệt độ trung bình theo tháng cho mỗi thành phố và hiển thị kết quả.
data.groupby('Thành_phố')['Lượng_mưa'].mean()
```

```python
# tạo bảng cột gồm 4 cột:
df = pd.DataFrame(columns=['Khu_vực', 'Nhiệt độ TB (°C)','Lượng mưa TB (mm)','AQI TB'])
df['Khu_vực'] = ['Bắc Bộ', 'Trung Bộ', 'Nam Bộ', 'Tây Nguyên']
df['Nhiệt độ TB (°C)'] = [data[data['Khu_vực'] == 'Bắc Bộ']['Nhiệt_độ'].mean(),
                        data[data['Khu_vực'] == 'Trung Bộ']['Nhiệt_độ'].mean(),
                        data[data['Khu_vực'] == 'Nam Bộ']['Nhiệt_độ'].mean(),
                         data[data['Khu_vực'] == 'Tây Nguyên']['Nhiệt_độ'].mean(),]
df['Lượng mưa TB (mm)'] = [data[data['Khu_vực'] == 'Bắc Bộ']['Lượng_mưa'].mean(),
                        data[data['Khu_vực'] == 'Trung Bộ']['Lượng_mưa'].mean(),
                        data[data['Khu_vực'] == 'Nam Bộ']['Lượng_mưa'].mean(),
                         data[data['Khu_vực'] == 'Tây Nguyên']['Lượng_mưa'].mean(),]
df['AQI TB'] = [data[data['Khu_vực'] == 'Bắc Bộ']['AQI'].mean(),
                        data[data['Khu_vực'] == 'Trung Bộ']['AQI'].mean(),
                        data[data['Khu_vực'] == 'Nam Bộ']['AQI'].mean(),
                         data[data['Khu_vực'] == 'Tây Nguyên']['AQI'].mean(),]

df
```

```python
# Từ đó trực quan hóa dữ liệu Bảng 1 bằng đồ thị Bar.
# quan hóa dữ liệu Bảng df bằng đồ thị Bar.
df.plot(kind='bar', x='Khu_vực', y=['Nhiệt độ TB (°C)','Lượng mưa TB (mm)','AQI TB'])
plt.show()
```
